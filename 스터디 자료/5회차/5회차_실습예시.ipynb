pandas 로 불러오기

```
data_pandas = pd.read_csv('ThoraricSurgery.csv',header=None) # 판다스가 좀 더 편하더라..
```

```
data_pandas
```

모듈 임포트 , 시드 설정

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)
```

x,y 데이터 분리

```
x_data = data_pandas.iloc[:,:17] # 판다스 인덱싱 
y_data = data_pandas.iloc[:,17]
x_data
```

```
y_data
```

## 모델 구조 결정
입력층 , 은닉층 , 출력층 설정

```
model = Sequential()
model.add(Dense(30,input_dim = 17,activation='relu')) # 30 개의 노드, 입력(x)의 차원 17,활성화함수 relu
model.add(Dense(1,activation='sigmoid')) # 출력층 , 활성화함수 sigmoid
```

##모델 실행
모델 컴파일

```
from tensorflow.keras import optimizers

optim = optimizers.Adam(learning_rate=0.002)  ## 옵티마이저 설정(경사하강법 등)
model.compile(optimizer = optim,loss='binary_crossentropy',metrics=['accuracy']) #손실함수를 교차엔트로피로 
```

## 모델 실행

epoch 와 batchsize 설명

```
model.fit(x_data,y_data,epochs=500,batch_size=8)
```

모델 평가

```
print("acc : %.2f"%(model.evaluate(x_data,y_data)[1]))
```

#### One-hot (원핫) 인코딩

encoding -> label encoding , one-hot encoding 등 설명 필요


```
from sklearn.preprocessing import LabelEncoder # 책에 있음
from sklearn.preprocessing import OneHotEncoder # 한번에 가능
```

```
# 여러 스케일러 
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import MinMaxScaler
```

```
x_data = data.iloc[:,:4]
y_data = data.iloc[:,4]
x_data
```

```
y_data
```

```
e = LabelEncoder()
e.fit(y_data)
Y_fit = e.transform(y_data)
Y_fit
```

```
from tensorflow.keras.utils import to_categorical # 책 오류
```

```
y_encoded = to_categorical(Y_fit)
y_encoded
```

이 과정을 한번에

```
one_hot = OneHotEncoder()
```

```
y_fited = one_hot.fit_transform(y_data.to_numpy().reshape(-1,1))
y_fited = y_fited.toarray()
y_fited
```

### 소프트 맥스(softmax)

모델 설정

```
model = Sequential()
model.add(Dense(16,input_dim=4,activation='relu')) # 입력층, 입력 x차원 4 
model.add(Dense(8,activation='relu')) # 은닉층
model.add(Dense(3,activation='softmax')) # 출력층 ,출력 3개(클래스 1,2,3) 소프트 맥스로
```

모델 컴파일

```
from tensorflow.keras import optimizers
optim = optimizers.Adam(lr=0.005)
model.compile(optimizer=optim,loss='categorical_crossentropy',metrics=['accuracy']) # 다중분류 이기 떄문  categorical_crossentropy 사용
```

모델 실행

```
model.fit(x_data,y_fited,epochs=50,batch_size=1)
```

모델 평가

```
print("acc : %.2f"%(model.evaluate(x_data,y_fited)[1]))
```
